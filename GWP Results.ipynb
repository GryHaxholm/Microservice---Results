```python
with open("/Users/gryhaxholm/Documents/DTU - Kandidatpeciale /Programming/Microservice - FEM_LCA_calc/resultsFiles/Beams_GWP.json") as b:
    dataBeams = json.load(b)
with open("/Users/gryhaxholm/Documents/DTU - Kandidatpeciale /Programming/Microservice - FEM_LCA_calc/resultsFiles/Columns_GWP.json") as c:
    dataColumns = json.load(c)
with open("/Users/gryhaxholm/Documents/DTU - Kandidatpeciale /Programming/Microservice - FEM_LCA_calc/resultsFiles/Decks_GWP.json") as d:
    dataDecks = json.load(d)
with open("/Users/gryhaxholm/Documents/DTU - Kandidatpeciale /Programming/Microservice - FEM_LCA_calc/resultsFiles/Foundations_GWP.json") as f:
    dataFoundations = json.load(f)
with open("/Users/gryhaxholm/Documents/DTU - Kandidatpeciale /Programming/Microservice - FEM_LCA_calc/resultsFiles/Walls_GWP.json") as w:
    dataWalls = json.load(w)
with open("/Users/gryhaxholm/Documents/DTU - Kandidatpeciale /Programming/Microservice - FEM_LCA_calc/resultsFiles/ExteriorWalls_GWP.json") as ew:
    dataExtWalls = json.load(ew)
with open("/Users/gryhaxholm/Documents/DTU - Kandidatpeciale /Programming/Microservice - FEM_LCA_calc/resultsFiles/InteriorWalls_GWP.json") as iw:
    dataIntWalls = json.load(iw)   
```

```python
df_Beams = pd.DataFrame(dataBeams, index=['Beams'])
df_Columns = pd.DataFrame(dataColumns, index=['Columns'])
df_Decks = pd.DataFrame(dataDecks, index=['Decks'])
df_Foundations = pd.DataFrame(dataFoundations, index=['Foundations'])
df_Walls = pd.DataFrame(dataWalls, index=['Walls'])
df_ExtWalls = pd.DataFrame(dataExtWalls, index=['ExteriorWalls'])
df_IntWalls = pd.DataFrame(dataIntWalls, index=['InteriorWalls'])
print(df_Foundations)
print(df_Beams)
print(df_Columns)
print(df_Decks)
print(df_Walls)
print(df_ExtWalls)
print(df_IntWalls)
```

```python
dfAll = [df_Beams,df_Columns, df_Decks,df_Foundations, df_Walls]
df_All = pd.concat(dfAll, join='outer')
df_All
```

```python
df_All_Pivot=pd.pivot_table(df_All, columns=['Beams','Columns','Decks', 'Foundations', 'Walls',], values=['A1-A3','C3','C4','D','Total'], aggfunc=np.sum)
print(df_All_Pivot)
```
